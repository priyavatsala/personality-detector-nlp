```python
import re
```

```python
def clean_text(text):
    # remove URLs
    text = re.sub(r'http\S+', '', text)
    
    # remove special characters
    text = re.sub(r'[^A-Za-z\s]', '', text)
    
    # convert to lowercase
    text = text.lower()
    
    return text
```

```python
df['cleaned_posts'] = df['posts'].apply(clean_text)
```
